```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/content/diabetes.csv')
```

```python
target=df['Outcome']
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
  if df[i].dtype!='object' and i!='Outcome':
    df[i]=df[i].replace(0,np.nan)
```

```python
df
```

```python
for i in df.columns:
  sns.boxplot(x=df[i])
  plt.show()
```

```python
df.isnull().sum()
```

```python
df[['Insulin','BMI']]=df[['Insulin','BMI']].fillna(df[['Insulin','BMI']].median())
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
  if df[i].dtype!='object':
    df[i]=df[i].fillna(df[i].mean())
```

```python
df
```

```python
for i in df.columns:
  sns.boxplot(x=df[i])
  plt.show()
```

```python
import random as r
def detout(column):
  q1=np.percentile(df[column],25)
  q3=np.percentile(df[column],75)
  iqr=q3-q1
  return q1-1.5*iqr,q3+1.5*iqr
for i in df.columns:
  if df[i].dtype!='object':
    lower,upper=detout(i)
    df[i]=np.where((df[i]>upper),upper,df[i])
    df[i]=np.where((df[i]<lower),lower,df[i])
```

```python
# sns.boxplot(df['BMI'])
```

```python
for i in df.columns:
  sns.boxplot(x=df[i])
  plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Outcome',axis=1),df['Outcome'],test_size=0.2,random_state=42)
```

```python
df
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,mean_squared_error,mean_absolute_error
y_pred_nb=nb.predict(X_test)
acc_nb=accuracy_score(y_test,y_pred_nb)
print(acc_nb)
mse_nb=mean_squared_error(y_test,y_pred_nb)
print(mse_nb)
mae_nb=mean_absolute_error(y_test,y_pred_nb)
print(mae_nb)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred_dt=dt.predict(X_test)
acc_dt=accuracy_score(y_test,y_pred_dt)
print(acc_dt)
mse_dt=mean_squared_error(y_test,y_pred_dt)
print(mse_dt)
mae_dt=mean_absolute_error(y_test,y_pred_dt)
print(mae_dt)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred_lr=lr.predict(X_test)
acc_lr=accuracy_score(y_test,y_pred_lr)
print(acc_lr)
mse_lr=mean_squared_error(y_test,y_pred_lr)
print(mse_lr)
mae_lr=mean_absolute_error(y_test,y_pred_lr)
print(mae_lr)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_pred_knn=knn.predict(X_test)
acc_knn=accuracy_score(y_test,y_pred_knn)
print(acc_knn)
mse_knn=mean_squared_error(y_test,y_pred_knn)
print(mse_knn)
mae_knn=mean_absolute_error(y_test,y_pred_knn)
print(mae_knn)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred_rf=rf.predict(X_test)
acc_rf=accuracy_score(y_test,y_pred_rf)
print(acc_rf)
mse_rf=mean_squared_error(y_test,y_pred_rf)
print(mse_rf)
mae_rf=mean_absolute_error(y_test,y_pred_rf)
print(mae_rf)
```

```python
# prompt: Hyperparameter tuning


import random as r
from sklearn.model_selection import train_test_split, GridSearchCV

# Hyperparameter tuning for Decision Tree
param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search_dt = GridSearchCV(DecisionTreeClassifier(), param_grid_dt, cv=5)
grid_search_dt.fit(X_train, y_train)
best_dt = grid_search_dt.best_estimator_
y_pred_best_dt = best_dt.predict(X_test)
print("Best Decision Tree Accuracy:", accuracy_score(y_test, y_pred_best_dt))

# Hyperparameter tuning for KNN
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}
grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5)
grid_search_knn.fit(X_train, y_train)
best_knn = grid_search_knn.best_estimator_
y_pred_best_knn = best_knn.predict(X_test)
print("Best KNN Accuracy:", accuracy_score(y_test, y_pred_best_knn))

# Hyperparameter tuning for Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search_rf = GridSearchCV(RandomForestClassifier(), param_grid_rf, cv=5)
grid_search_rf.fit(X_train, y_train)
best_rf = grid_search_rf.best_estimator_
y_pred_best_rf = best_rf.predict(X_test)
print("Best Random Forest Accuracy:", accuracy_score(y_test, y_pred_best_rf))

```
